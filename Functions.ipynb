* Map() - Executes a specified function for each of item of an iterable.

       num_as_strings = ["2","4","6","8"]

         from functools import reduce
         num = [1,2,3,4]
         total_sum = reduce(lambda( x,y:x+y, num))
         print(total_sum)
         #output: 10

* Filter() - It is used to filter elements from an iterable based on some condition.
     Example:
        l1 =[-1,-2,3,4,5]
        list(filter(lambda x: x<0,l1))

```python
#1 Write a python function that takes a list of numbers as input and returns the sum of all even numbers in list.

def sum_even_numbers(num):
  sum =0
  for numbers in num:
    if numbers%2==0:
       sum += numbers
  return sum
sum_even_numbers([2,3,5,8,9,12,13,14])
```

```python
#4 Write a python function that checks if a given number is prime or not from 1 to 200.

def is_prime_num(n):
  for num in range(1,201):
    if n<=1:
      return False
    for i in range(2,int(n**0.5)+1):
      if n%i==0:
        return False
    return True
is_prime_num(197)
```

```python
# 5 Create an iterator class in python that generates the fibonacci sequence upto a specified number of terms.

def fib_generator():
    a=0
    b=1
    while True:
      yield a
      a,b = b,a+b
gen = fib_generator()
for i in range(10):
  print(next(gen))
```

```python
#6 Write a generator function that yields the power of 2 upto a given exponent.

def power_of_two(n):
    for i in range(0,n):
      yield 2**i
gen = power_of_two(6)
list(gen)
```

```python
#7 Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as f:
            for line in f:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")


with open('my_test_file.txt', 'w') as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("And this is the third line.")

file_lines_generator = read_file_line_by_line('my_test_file.txt')

for line in file_lines_generator:
    print(line)
```

```python
#9 Write a python program that uses 'map()' to convert a list of temperature from Celsius to Fahrenheit.

temp_in_Celsius = [30,45,60,97]
temp_in_Fahrenheit = list(map(lambda Celsius : Celsius*9/5+32,temp_in_Celsius))
print(temp_in_Fahrenheit)
```
